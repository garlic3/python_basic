```python
import selenium 
from selenium import webdriver
from bs4 import BeautifulSoup as bs

URL = 'https://www.genie.co.kr/chart/top200'
driver = webdriver.Chrome('chromedriver.exe')
driver.get(URL)
html = driver.page_source
soup = bs(html)
songs = soup.select('tbody > tr')
song_data = []
rank = 1
for song in songs:
    title = song.select('a.title')[0].text.replace('\n','')
    singer = song.select('a.artist')[0].text.replace('\n','')
    song_data.append([rank, title.strip(), singer.strip()])
    rank += 1
print(song_data)

import pandas as pd
df = pd.DataFrame(song_data, columns = ['순위','타이틀','가수'])
df.to_excel('genie.xlsx', index=False)
```
