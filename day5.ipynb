알고리즘
- 문제나 과제를 해결하기 위한 처리 절차를 구체적인 순서로 표현한 아이디어 또는 생각

알고리즘과 프로그램
- 알고리즘은 프로그래밍 언어를 기술하면 프로그램

프로그램 작성에서의 알고리즘
- 프로그램 작성은 기획, 설계, 프로그래밍, 디버그로 진행
- 설게 단계에서 알고리즘 필요

좋은 알고리즘?
- 이해하기 쉬움, 속도가 빠름, 효율적(메모리를 작게 사용), 재이용이 쉬움(객체지향프로그래밍의 특징)

왜 알고리즘을 공부해야 하나
- 좋은 프로그램을 만들기 위해(빠르고, 효율적, 범용성)
- 프로그램의 좋고 나쁨을 판단하기 위해
- 프로그램 작성 과정 자체를 효율화하기 위해
- 프로그램 기술을 향상시키기 위해

알고리즘의 조건
- 정확한 결과를 얻을 수 있어야 한다
- 반드시 종료되어야 한다(중간 처리나 절차가 잘못 만들어지면 영원히 끝나지 않음)

알고리즘의 기본형 3가지
- 순차구조: 처음부터 순서대로 처리하는 절차
  예) 심부름 
    1. 슈퍼마켓에 간다
    2. 라면을 산다
    3. 집으로 돌아온다
  실랭을 원하는 처리를 위부터 순서대로 작성한다 알고리즘 중에섣소 가장 많이 사용된다
- 선택구조: 조건식으로 판단해 실행할 절차를 전환하는 구조
- 반복구조: 조건을 만족하는 동안 같은 처리를 반복하는 절차

```python
def sum_all(start,end):
  output = 0
  for i in range(start, end+1):
    output = output + i
  return output
print('0에서 100까지의 합:',sum_all(0,100))  
print('0에서 100까지의 합:',sum_all(0,1000))
print('0에서 100까지의 합:', sum_all(50,100))
print('0에서 100까지의 합:', sum_all(500,1000))
```

```python
# 재귀함수
def fac(n):
  if n == 0:
    return 1
  else:
    return n*fac(n-1)

print('3!:',fac(3))
print('5!:',fac(5))
print('7!:', fac(7))
```

```python
def facto(n):
  output = 1
  for i in range(1, n+1):
    output *= i
  return output

print('3!:',facto(3))
print('5!:',facto(5))
print('7!:', facto(7))
```

```python
# 피보나치 수열(토끼의 번식속도, 바이러스 감염 속도)

def fibo(n): 
  if n < 3:
    return 1
  else:
    output = 1
    output = fibo(n-1)+ fibo(n-2)
    return output

print('fibonacci(1):',fibo(1))
print('fibonacci(2):',fibo(2))
print('fibonacci(3):',fibo(3))
print('fibonacci(4):',fibo(4))
print('fibonacci(5):',fibo(5))

```

```python
# 튜플과 함수
# 튜플은 함수의 리턴에 많이 사용된다. 함수의 리턴에 튜플을 사용하게 되면 여러개의 값을
# 리턴하고 할당 할수 있기 때문이다. 

def test():
  return 10, 20  # 튜플의 (소)괄호는 생략도 가능하다. 

a, b = test()

print(a)
print(b)
```
